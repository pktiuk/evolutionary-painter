```python
b = BindWrapper()
```

```python
#Set parameters of model
b.filename = "figury_biale.png"
b.size = 30
b.genes_count = 100
b.max_generation = 2000
b.has_window = False
b.save_result_image = False
b.execution_timeout = 0
```

```python
multiple_sizes_data = b.test_multiple_times("size",min_value=10,number_of_steps=5,step_size=20,number_of_tries=2)
generations = multiple_sizes_data.pop("gen")
```

```python
for key, value in multiple_sizes_data.items():
    plt.plot(range(len(value), value, label=str(key))
plt.legend(title="Size")
plt.title("Multiple sizes of population for " + b.filename)
plt.rcParams["figure.figsize"]=15,8

plt.show()
```
